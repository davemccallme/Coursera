```python
def parse_line(line):
    row = json.loads(line)
    df = pd.DataFrame.from_dict(row, orient='index').transpose()
    if not df.empty and not df['Name'].iloc[0][0].isdigit():
        print(f"Skipping header row: {df}")
        return None  # skip header rows
    return df



def extract(file_path: str, columns: List[str]) -> pd.DataFrame:
    extract_json = pd.DataFrame(columns=columns)
    with open(file_path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            row = json.loads(line)
            df = pd.DataFrame.from_dict(row, orient='index').transpose()
            if not df.empty and df['Name'].iloc[0].isdigit():
                extract_json = pd.concat([extract_json, df], ignore_index=True)
    return extract_json

```

## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1.  Changes the `Market Cap (US$ Billion)` column from USD to GBP
2.  Rounds the Market Cap (US$ Billion)\` column to 3 decimal places
3.  Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python

def transform(df: pd.DataFrame, exchange_rate: pd.Series) -> pd.DataFrame:
    # Convert Market Cap from USD to GBP
    df['Market Cap (GBP$ Billion)'] = df['Market Cap (US$ Billion)'] * exchange_rate
    
    # Drop Market Cap (US$ Billion) column
    df = df.drop('Market Cap (US$ Billion)', axis=1)
    
    # Rename columns
    df = df.rename(columns={
        'Name': 'Code',
        'Market Cap (GBP$ Billion)': 'Market Cap (Billion GBP)',
    })
    
    # Set index to Code column
    df = df.set_index('Code')
    
    return df

df = transform(extract(file_to_process, columns), exchange_rate)

```
